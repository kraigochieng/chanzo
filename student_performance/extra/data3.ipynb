```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from faker import Faker
from typing import Literal
```

```python
fake = Faker()

grades = ["A", "B", "C", "D", "E"]

grade_levels = ["Good", "Mid", "Bad"]

grade_weights_per_grade_level = {
    "Good": [0.6, 0.3, 0.08, 0.02, 0.0],  # Mostly A and B
    "Mid": [0.2, 0.3, 0.3, 0.1, 0.1],  # Even spread
    "Bad": [0.05, 0.1, 0.25, 0.3, 0.3],  # More D and E
}


subject_tuples = [
    ("Languages", "English", "English Reading"),
    ("Languages", "English", "English Writing"),
    ("Languages", "Spanish", "Vocabulary"),
    ("Languages", "Spanish", "Spanish Grammar"),
    ("Sciences", "Math", "Algebra"),
    ("Sciences", "Math", "Geometry"),
    ("Sciences", "Physics", "Mechanics"),
    ("Sciences", "Physics", "Thermodynamics"),
    ("Humanities", "Geography", "Physical Geography"),
    ("Humanities", "Geography", "Human Geography"),
    ("Humanities", "History", "Ancient History"),
    ("Humanities", "History", "Medieval History"),
]



terms = list(range(1, 4))
```

```python
def generate_student_df(
    grade_level: Literal["Good", "Bad", "Mid"],
    stream: str,
    name: str = None,
) -> pd.DataFrame:
    if name is None:
        name = fake.first_name()

    data = []

    for subject_category, subject, subject_strand in subject_tuples:
        for term in terms:
            grade = random.choices(
                population=grades, weights=grade_weights_per_grade_level[grade_level]
            )[0]

            data.append(
                [stream, name, subject_category, subject, subject_strand, term, grade]
            )

    # Convert to DataFrame
    return pd.DataFrame(
        data,
        columns=[
            "Stream",
            "Student Name",
            "Subject Category",
            "Subject",
            "Subject Strand",
            "Term",
            "Grade",
        ],
    )
```
